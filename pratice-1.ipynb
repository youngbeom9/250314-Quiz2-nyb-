```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
act = pd.read_csv(r"C:\Users\302-28\Downloads\OTg6QzA_activities.csv")
```

```python
X = act.X.to_numpy()
Y = act.Y.to_numpy()
```

```python
fig=plt.figure(figsize=(18,12))

fig.add_subplot(2,2,1)
plt.hist2d(X[:16570], Y[:16570], bins=20, cmap='viridis')
plt.title('1-Q')
plt.colorbar()

fig.add_subplot(2,2,2)
plt.hist2d(X[16570:16570*2], Y[16570:16570*2], bins=20, cmap='gray')
plt.title('2-Q')
plt.colorbar()

fig.add_subplot(2,2,3)
plt.hist2d(X[16570*2:16570*3], Y[16570*2:16570*3], bins=20, cmap='magma')
plt.title('3-Q')
plt.colorbar()

fig.add_subplot(2,2,4)
plt.hist2d(X[16570*3:], Y[16570*3:], bins=20, cmap='magma')
plt.title('4-Q')
plt.colorbar()

plt.show()
```

```python
fig=plt.figure(figsize=(18,12),facecolor='gray')

fig.add_subplot(2,2,1)
plt.scatter(X[:16570], Y[:16570],c='k')
plt.title('1-Q')

fig.add_subplot(2,2,2)
plt.scatter(X[16570:16570*2], Y[16570:16570*2],c='k')
plt.title('2-Q')

fig.add_subplot(2,2,3)
plt.scatter(X[16570*2:16570*3], Y[16570*2:16570*3],c='k')
plt.title('3-Q')

fig.add_subplot(2,2,4)
plt.scatter(X[16570*3:], Y[16570*3:],c='k')
plt.title('4-Q')

plt.show()
```
